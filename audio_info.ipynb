```python
import os
import pandas as pd
import copy
```

```python
filesnames = []
with os.scandir('archive/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files') as entries:
    for entry in entries:
        if entry.name.endswith('.txt'):
            filesnames.append(entry.name)
```

```python
print(len(filesnames))
```

```python
patient_data = {}
```

```python
patient_number = []
recording_equipment = []

percent_wheeze_Al = []
percent_cracking_Al = []

percent_wheeze_Ar = []
percent_cracking_Ar = []

percent_wheeze_Pl = []
percent_cracking_Pl = []

percent_wheeze_Pr = []
percent_cracking_Pr = []

percent_wheeze_Ll = []
percent_cracking_Ll = []

percent_wheeze_Lr = []
percent_cracking_Lr = []
```

```python
#dict contains all files corresponding to each patient
for filename in filesnames:
    if filename.split('_')[0] not in patient_data.keys():
        patient_data[filename.split('_')[0]] = []
    patient_data[filename.split('_')[0]].append(filename)
```

```python
def get_cw_data(filename):
    lines = []
    
    with open(filename) as f:
        lines = f.readlines()
    lines = [x.strip().split('\t') for x in lines]

    cracks_total = 0
    wheezing_total = 0
    num_cycles = len(lines)
    
    for line in lines:
        cracks_total += int(line[2])
        wheezing_total += int(line[3])
    
    return cracks_total/num_cycles, wheezing_total/num_cycles
```

```python
for patient in patient_data.keys():
    num_experiments = len(patient_data[patient])
    
    patient_number.append(patient)
    cur_recording_equipment = ''
    
    total_wheeze_Al = 0
    total_cracking_Al = 0
    num_Al = 0

    total_wheeze_Ar = 0
    total_cracking_Ar = 0
    num_Ar = 0

    total_wheeze_Pl = 0
    total_cracking_Pl = 0
    num_Pl = 0

    total_wheeze_Pr = 0
    total_cracking_Pr = 0
    num_Pr = 0

    total_wheeze_Ll = 0
    total_cracking_Ll = 0
    num_Ll = 0

    total_wheeze_Lr = 0
    total_cracking_Lr = 0
    num_Lr = 0

    for recording in patient_data[patient]:
        cur_path = f'archive/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files/{recording}'
        
        data = recording[:-4].split('_')
        cur_recording_equipment =  data[-1]
        
        cur_loc = data[2]

        if cur_loc == 'Al':
            num_Al += 1
            cw_data = get_cw_data(cur_path)

            total_cracking_Al += cw_data[0]
            total_wheeze_Al += cw_data[1]

        elif cur_loc == 'Ar':
            num_Ar += 1
            cw_data = get_cw_data(cur_path)

            total_cracking_Ar += cw_data[0]
            total_wheeze_Ar += cw_data[1]
        
        elif cur_loc == 'Pl':
            num_Pl += 1
            cw_data = get_cw_data(cur_path)

            total_cracking_Pl += cw_data[0]
            total_wheeze_Pl += cw_data[1]
            
        elif cur_loc == 'Pr':
            num_Pr += 1
            cw_data = get_cw_data(cur_path)

            total_cracking_Pr += cw_data[0]
            total_wheeze_Pr += cw_data[1]
        
        elif cur_loc == 'Ll':
            num_Ll += 1
            cw_data = get_cw_data(cur_path)

            total_cracking_Ll += cw_data[0]
            total_wheeze_Ll += cw_data[1]
            
        elif cur_loc == 'Lr':
            num_Lr += 1
            cw_data = get_cw_data(cur_path)

            total_cracking_Lr += cw_data[0]
            total_wheeze_Lr += cw_data[1]
    
    recording_equipment.append(cur_recording_equipment)
    
    percent_wheeze_Al.append(total_wheeze_Al/num_Al if num_Al > 0 else None)
    percent_cracking_Al.append(total_cracking_Al/num_Al if num_Al > 0 else None)

    percent_wheeze_Ar.append(total_wheeze_Ar/num_Ar if num_Ar > 0 else None)
    percent_cracking_Ar.append(total_cracking_Ar/num_Ar if num_Ar > 0 else None)

    percent_wheeze_Pl.append(total_cracking_Pl/num_Pl if num_Pl > 0 else None)
    percent_cracking_Pl.append(total_cracking_Pl/num_Pl if num_Pl > 0 else None)

    percent_wheeze_Pr.append(total_cracking_Pr/num_Pr if num_Pr > 0 else None)
    percent_cracking_Pr.append(total_cracking_Pr/num_Pr if num_Pr > 0 else None)

    percent_wheeze_Ll.append(total_cracking_Ll/num_Ll if num_Ll > 0 else None)
    percent_cracking_Ll.append(total_cracking_Ll/num_Ll if num_Ll > 0 else None)

    percent_wheeze_Lr.append(total_cracking_Lr/num_Lr if num_Lr > 0 else None)
    percent_cracking_Lr.append(total_cracking_Lr/num_Lr if num_Lr > 0 else None)
```

```python
data = {'patient_number': patient_number,
    'recording_equipment': recording_equipment,
    'percent_wheeze_Al': percent_wheeze_Al,
    'percent_cracking_Al': percent_cracking_Al,
    'percent_wheeze_Ar': percent_wheeze_Ar,
    'percent_cracking_Ar': percent_cracking_Ar,
    'percent_wheeze_Pl': percent_wheeze_Pl,
    'percent_cracking_Pl': percent_cracking_Pl,
    'percent_wheeze_Pr': percent_wheeze_Pr,
    'percent_cracking_Pr': percent_cracking_Pr,
    'percent_wheeze_Ll': percent_wheeze_Ll,
    'percent_cracking_Ll': percent_cracking_Ll,
    'percent_wheeze_Lr': percent_wheeze_Lr,
    'percent_cracking_Lr': percent_cracking_Lr}

df = pd.DataFrame(data)
df.head()
```

```python
df_median = copy.deepcopy(df)
df_mean = copy.deepcopy(df)
```

```python
cols_to_fill = [col for col in df_median.columns if col not in ['recording_equipment', 'patient_number']]
df_median[cols_to_fill] = df_median[cols_to_fill].fillna(df_median[cols_to_fill].median())
df_median['patient_number'] = df_median['patient_number'].astype('int64')
df_median.head()
```

```python
cols_to_fill = [col for col in df_mean.columns if col not in ['recording_equipment', 'patient_number']]
df_mean[cols_to_fill] = df_mean[cols_to_fill].fillna(df_mean[cols_to_fill].mean())
df_mean['patient_number'] = df_mean['patient_number'].astype('int64')
df_mean.head()
```

```python
df.to_csv('audio_info.csv', index=True)
df_median.to_csv('audio_info_median.csv', index=True)
df_median.to_csv('audio_info_median.csv', index=True)
```
