```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
healthy_102_full, healthy_102_full_sr = librosa.load('archive/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files/102_1b1_Ar_sc_Meditron.wav')
healthy_102_cycle, healthy_102_cycle_sr = librosa.load('archive/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files/102_1b1_Ar_sc_Meditron.wav', offset=0.264, duration=1.472)
```

```python
librosa.display.waveshow(healthy_102_full)
plt.title('Patient 102 Full Clip')
plt.savefig('patient_102_full.png')
plt.show()

librosa.display.waveshow(healthy_102_cycle)
plt.title('Patient 102 Single Cycle')
plt.savefig('patient_102_cycle.png')
plt.show()
```

```python
healthy_102_full_spectrogram = librosa.feature.melspectrogram(y = healthy_102_full, sr = healthy_102_full_sr)
healthy_102_cycle_spectrogram = librosa.feature.melspectrogram(y = healthy_102_cycle, sr = healthy_102_cycle_sr)
```

```python
healthy_102_full_spectrogram_dB = librosa.power_to_db(healthy_102_full_spectrogram, ref=np.max)
healthy_102_cycle_spectrogram_dB = librosa.power_to_db(healthy_102_cycle_spectrogram, ref=np.max)
```

```python
healthy_102_full_mfcss = librosa.feature.mfcc(S=healthy_102_full_spectrogram_dB, n_mfcc=20)
healthy_102_cycle_mfcss = librosa.feature.mfcc(S=healthy_102_cycle_spectrogram_dB, n_mfcc=20)
```

```python
fig, ax = plt.subplots(nrows=2, sharex=True)
img = librosa.display.specshow(healthy_102_full_spectrogram_dB, x_axis='time', y_axis='mel', ax=ax[0])
fig.colorbar(img, ax=[ax[0]])
ax[0].set(title='Mel-frequency spectrogram')
ax[0].label_outer()
img = librosa.display.specshow(healthy_102_full_mfcss, x_axis='time', ax=ax[1])
fig.colorbar(img, ax=[ax[1]])
ax[1].set(title='MFCC')
fig.suptitle('Patient 102, healthy: no crackles/no wheezing [Full Clip]')
```

```python
fig, ax = plt.subplots(nrows=2, sharex=True)
img = librosa.display.specshow(healthy_102_cycle_spectrogram_dB, x_axis='time', y_axis='mel', ax=ax[0])
fig.colorbar(img, ax=[ax[0]])
ax[0].set(title='Mel-frequency spectrogram')
ax[0].label_outer()
img = librosa.display.specshow(healthy_102_cycle_mfcss, x_axis='time', ax=ax[1])
fig.colorbar(img, ax=[ax[1]])
ax[1].set(title='MFCC')
fig.suptitle('Patient 102, healthy: no crackles/no wheezing [Single Cycle]')
```
